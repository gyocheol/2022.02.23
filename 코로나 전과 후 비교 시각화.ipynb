```python
data = pd.DataFrame({
    '평균' : df['평균'],
    'before & after' : 'before'
})
```

```python
data.iloc[21:, 1] = 'after'
```

```python
covid = pd.merge(df, data, on='평균')
```

```python
covid.drop_duplicates()
```

```python
covid = covid.drop([covid.index[8],covid.index[15], covid.index[16], covid.index[2], covid.index[10], covid.index[9]])
```

```python
covid = covid.reset_index()
covid
```

```python
covid.drop('index', 1)
```

```python
plt.figure(figsize=(15, 10))

splot = sns.barplot(data=covid, x='하위상권', y='평균', hue='before & after')
plt.show()
sfig = splot.get_figure()
sfig.savefig('코로나 전과 후 공실률.png')
```
